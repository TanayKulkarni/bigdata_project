```python
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient()
```

```python
mydb = client['covid']
covid = mydb['timeseries']
all_records = mydb['all_records']
```

```python
for x in covid.find():
    print(x)
```

```python
df = pd.DataFrame(list(covid.find()))
```

```python
df = df[['location','total_cases','total_deaths']]
```

```python
df.to_excel('total.xlsx')
```

```python
pd.DataFrame(list(all_records.find())).drop(columns = ['_id']).to_excel('all_records.xlsx')
```
